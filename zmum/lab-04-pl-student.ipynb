# Potencjalny problem z algorytmem lasso, tj. karą $\mathcal{l}_1$

Hui Zhou [The adaptive lasso and its oracle properties](https://www.tandfonline.com/doi/abs/10.1198/016214506000000735):

"Lasso variable selection has been shown to be consistent under certain conditions. In this work we derive a necessary condition for the lasso variable selection to be consistent. Consequently, there exist certain scenarios where the lasso is inconsistent for variable selection."

*Consistency* tłumaczymy jako *zgodność*.

Problem minimizacji z karą adaptive lasso

\begin{equation}
\hat{\beta}_\text{ada} = \underset{\beta}{\text{arg min}} \: L(\beta; x, y)
+ \lambda \sum_{j=1}^p \hat{w}_j | \beta_j |
\end{equation}

Metoda *adaptive lasso* działa także dla uogólnionych modeli liniowych, w szczególności: dla regresji logistycznej.
